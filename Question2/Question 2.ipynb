```python
#Using Sachin tendulkars last speech as corpus
#Create a list of all characters by index and store into a list
f = open('sachin.txt', 'r')
list_char = []
while True:
    ch=f.read(1)
    if not ch: break
    list_char.append(ch.lower())
print list_char
        
```

```python
#Input a string and pull out the last character
inp = raw_input("Enter a character or word to predict next letter")
letter = inp[-1]
```

```python
#Create a list of letter at position index + 1
next_elem = []
indices = [i for i, x in enumerate(list_char) if x == letter]
for index in indices:
    next_elem.append(list_char[index+1])
print next_elem    

        
```

```python
#Using the Collections Library to easily find out top three letter
from collections import Counter
high= [ite for ite, it in Counter(next_elem).most_common(3)]
```

```python
#Print the predicted letters to the user
print("The next three predicted outputs are")
print "1",high[0],"\n","2",high[1],"\n","3",high[2]
```
